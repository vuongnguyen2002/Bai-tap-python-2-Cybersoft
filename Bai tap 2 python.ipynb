```python
import csv
listthanhpho=[]
listdsnamnu =[]
tong_don =[]
with open('example_bai_1.2.csv', 'r',encoding='utf-8') as f:
    data_reader = csv.reader(f)
   
    print(type(data_reader))
    for row in data_reader:
        #print(row)
        listthanhpho.append(row[0])
        listdsnamnu.append(row[1])
        tong_don.append(row[2])
print(f"thanh pho la:" ,listthanhpho)
#print( "Danh sach nam nu la : ",listdsnamnu)
print("tong don la: ",tong_don)
```
